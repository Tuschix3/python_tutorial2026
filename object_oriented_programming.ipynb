# Klassen / *Classes*

Eine Klasse in Python ist ein Bauplan oder eine Vorlage, die verwendet wird, um Objekte zu erstellen, und definiert die Attribute (Daten) sowie Methoden (Funktionen), die die Objekte besitzen werden.

---

*A class in Python is a blueprint or template used to create objects, defining the attributes (data) and methods (functions) that the objects will have.*

````python
class Player:
    def __init__(self, name, health, attack):
        self.name = name
        self.health = health
        self.attack = attack
````

### Objekte / *Objects*

Ein Objekt ist eine Instanz einer Klasse, die die tatsächlichen Daten (Attribute) enthält und die in der Klasse definierten Methoden verwenden kann.

---

*An object is an instance of a class that contains the actual data (attributes) and can use the methods defined in the class.*

````python
# creating players
red_beard = Player("Red Beard", 100, 20)
dark_smoke = Player("Darksmoke", 1000, 40)

print(red_beard.name, red_beard.health)
print(dark_smoke.name, dark_smoke.health)
````

### self keyword

In Python bezieht sich `self` auf die aktuelle Instanz (das Objekt) einer Klasse und wird verwendet, um innerhalb der Klasse auf die Attribute und Methoden des Objekts zuzugreifen.

---

*In Python, `self` refers to the current instance (object) of a class and is used to access the object's attributes and methods inside the class.*

### Funktionen vs. Methoden / *Functions vs Methods*

Eine Funktion ist ein eigenständiger Codeblock und benötigt nicht unbedingt eine Objektinstanz, um aufgerufen zu werden, während eine Methode eine Art von Funktion ist, die mit einem Objekt verknüpft ist und auf dieses Objekt wirkt.

Zum Beispiel ist `print()` eine eingebaute Funktion und funktioniert unabhängig von Klassen. Aber `append()` ist eine Methode, und  ein Objekt vom Typ `list` wird benötigt, um sie aufzurufen.

---

*A function is a standalone block of code and does not necessarily need an object instance to be called, while a method is a type of function that is associated with and operates on an object.*

*For example, `print()` is a built-in function and it works independently of classes. But `append()` is a method and you would need a `list` type object to call it.*

````python
numbers = [1,2,3]
print(numbers)
print(len(numbers))
````

# Vererbung / *Inheritance*

Was ist, wenn wir verschiedene Spielertypen haben wollen, die unterschiedliche Funktionen ausführen?

Zum Beispiel ein:e Zauberer:in, der/die zusätzlich einen Zauber wirken kann, um anzugreifen, ein:e Heiler:in, der/die heilen und die Gesundheit eines anderen Spielers/einer anderen Spielerin erhöhen kann, usw.

---

*What if we want different types of players who functions differently?*

*For example, a Wizard who can also cast a spell to attack, a healer who can heal and increase the health of another player, etc.*

````python
# define wizard
# a wizard can cast a spell and each spell cost 20 health points but each spell extracts 60 points from enemy

class Wizard(Player):
    def __init__(self, name, health, attack):
      #Player.__init__
        super().__init__(name, health, attack)
        self.spell = {
            "cost": 20, "harm": 60
        }

    def __str__(self):
        return "This player is a wizard and can cast a spell."

    def cast_spell(self, enemy):
        print(f"{self.name} is casting spell on {enemy.name}")
        self.health -= self.spell["cost"]
        enemy.health -= self.spell["harm"]
        print(f"{enemy.name}'s health is now {enemy.health}")
        print(f"{self.name}'s health is now {self.health}")
````

````python
merlin = Wizard("Merlin", 200, 20)
print(merlin)
merlin.attack_enemy(red_beard)
print(red_beard)
print(merlin.health)
merlin.cast_spell(red_beard)
print(red_beard)
print(merlin.health)
````

# Zugriffsmodifikatoren / *Access Specifiers*

In der Programmierung werden Zugriffsmodifikatoren verwendet, um die Sichtbarkeit und Zugänglichkeit von Klassenattributen und -methoden zu steuern. Im Gegensatz zu einigen anderen Sprachen erzwingt Python diese Zugriffsebenen jedoch nicht strikt, sondern verwendet Namenskonventionen, um anzuzeigen, wie Mitglieder verwendet werden sollten.

| Typ        | Syntax-Beispiel | Bedeutung                                         |
| ---------- | --------------- | ------------------------------------------------- |
| Öffentlich | `name`          | Von überall zugänglich                            |
| Geschützt  | `_name`         | Für den internen Gebrauch vorgesehen (Konvention) |
| Privat     | `__name`        | Stark eingeschränkt (Name Mangling)               |


---

*In programming, access specifiers are used to control the visibility and accessibility of class attributes and methods. But unlike some other languages, Python does not strictly enforce these access levels but uses naming conventions to indicate how members should be accessed.*

| Type      | Syntax Example  | Meaning                                |
| --------- | --------------- | -------------------------------------- |
| Public    | `name`          | Accessible from anywhere               |
| Protected | `_name`         | Intended for internal use (convention) |
| Private   | `__name`        | Strongly restricted (name mangling)    |


````python
class Player:
    def __init__(self, name, health, attack):
        self.name = name            # public
        self.health = health        # public
        self.attack = attack        # public
        self.is_alive = True        # public

        self.__max_health = health  # private
        self._level = 1             # Protected
        self._experience = 0        # Protected

    def __str__(self):
        if self.check_if_alive():
            return f"Player details:\nName: {self.name}, Health: {self.health}\nAn attack by {self.name} cause damage of {self.attack} health points."
        else:
            return f"Player {self.name} is dead."

    def check_if_alive(self):
        self.is_alive = self.health > 0
        return self.is_alive

    def attack_enemy(self, enemy):
        # enemy is a 'Player' object
        print(f"{self.name} is attacking {enemy.name}")
        enemy.health -= self.attack
        print(f"{enemy.name} lost {self.attack} health points")

    def get_max_health(self):
        return self.__max_health
````
