## Python Modules

Ein Modul in Python ist eine Datei, die Funktionen, Variablen oder Klassen enthält, die in anderen Programmen wiederverwendet werden können. Es hilft, Code zu organisieren und Wiederholungen zu vermeiden. Wenn der Programm an Größe zunimmt, wollen wir ihn möglicherweise in mehrere Dateien aufteilen, die üblicherweise als `module` bezeichnet werden. Dies erleichtert die Wartung und verbessert die Wiederverwendbarkeit des Codes. Wir können die am häufigsten verwendeten Funktionen in einem Modul definieren und es importieren, anstatt ihre Definitionen in verschiedene Programme zu kopieren. Ein Modul kann von einem anderen Programm importiert werden, um dessen Funktionalität zu nutzen. So werden auch die Python-Standardbibliotheken verwendet.

---

*A module in Python is a file that contains functions, variables, or classes that can be reused in other programs. It helps organize code and avoid repetition. As your program grows in size, you may want to split it into several files, commonly referred to as `modules`, for easier maintenance as well as reusability of the code. You can define your most used functions in a module and import it, instead of copying their definitions into different programs. A module can be imported by another program to make use of its functionality. This is how you use the Python standard libraries as well.*

### Verwendung integrierter Module / *Using Built-in Modules*

Python stellt viele integrierte Module wie `math`, `datetime`, `random` und so weiter bereit.

---

*Python provides many built-in modules like `math`, `datetime`, `random` and so on.*

````python
import math

print(math.sqrt(16))
print(math.pi)
````

````python
from math import sqrt

print(sqrt(25))
````

````python
import datetime

now = datetime.datetime.now()
print(now)
````

````python
import random

print(random.randint(1, 10))
print(random.choice(["apple", "banana", "cherry"]))
````

### Benutzerdefiniertes Modul / *Custom Module*

Um ein Modul zu erstellen, speichere einfach den Code in einer Datei mit der Dateiendung `.py`. Schauen wir uns `math_tools.py` an. Wir können dieses Modul verwenden, indem wir die Import-Anweisung benutzen. Zum Beispiel importieren wir die Funktion `add` und verwenden sie.

---

*To create a module just save the code you want in a file with the file extension `.py`. Lets look at `math_tools.py`. We can use this module, by using the import statement. For example, lets import the `add` function and use it.*

````python
x = 98
y = 45
````

````python
from modules.math_tools import add

sum_xy = add(x, y)
print(sum_xy)
````

Aber können wir `multiply`, eine weitere Funktion aus demselben Modul, verwenden?

---

*But can we use `multiply`, another function from the same module?*

````python
product_xy = multiply(x, y)
print(product_xy)
````

Was ist, wenn wir den vollständigen Pfad hinzufügen?

---

*What if we add the complete path?*

````python
product_xy = modules.math_tools.multiply(x, y)
print(product_xy)
````

Um alle im Modul definierten Funktionen zu importieren, können wir die folgenden Anweisungen verwenden:

---

*To import every functions defined in the module, we can use the following statements:*

````python
from modules.math_tools import add, subtract, multiply, divide

difference = subtract(x, y)
print(difference)

product_xy = multiply(x, y)
print(product_xy)

r_xy = divide(x, y)
print(r_xy)
````

Das Auflisten aller Funktionen kann jedoch mühsam sein. Stattdessen können wir einfach Folgendes verwenden.

---

*However, listing all the functions can be tedious. Instead we can just use the following.*

````python
from modules.math_tools import *

difference = subtract(x, y)
print(difference)

product_xy = multiply(x, y)
print(product_xy)

r_xy = divide(x, y)
print(r_xy)
````

````python
import modules.math_tools

sum_xy = modules.math_tools.add(x, y)
print(sum_xy)

product_xy = modules.math_tools.multiply(x, y)
print(product_xy)
````

````python
import modules.math_tools as mtools

sum_xy = mtools.add(x, y)
print(sum_xy)

product_xy = mtools.multiply(x, y)
print(product_xy)
````
