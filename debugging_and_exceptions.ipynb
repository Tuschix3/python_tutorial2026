# Edge Case

Betrachten wir die zuvor besprochene Funktion `divide`:
```
def divide(x, y):
    print(x / y)
```

Sie liefert das korrekte Ergebnis, außer wenn `y = 0` ist. Daher ist `y = 0` ein Edge Case für diese Funktion.

---

*An edge case is an unusual or extreme input value that occurs at the boundary of a program’s expected input range and may reveal bugs or unexpected behavior.*

*Consider the `divide` function we saw before:*
```
def divide(x, y):
    print(x / y)
```

*It returns correct result except when `y=0`. So `y = 0` is an edge case for this function.*

# Eingabevalidierung / *Input Validation*

Eingabevalidierung ist der Prozess, bei dem überprüft wird, ob Benutzereingaben gültig sind, bevor sie in einem Programm verwendet werden. Sie hilft, Fehler und unerwartetes Verhalten zu vermeiden und stellt sicher, dass das Programm Daten im richtigen Format oder Bereich erhält.

Einige Beispiele:
- Fehlende oder leere Eingabewerte in Pflichtfeldern, z. B. leere Strings, wenn die Länge des Strings kritisch ist
- Negative Werte, wo nur positive Werte akzeptiert werden, z.B. -15 als Menge einer Zutat in einem Rezept
- Werte, die zu klein oder zu groß sind, z. B. bei der Arbeit mit Datums- und Zeitangaben
- Ungültige Indizes für Sequenzen, z. B. der Versuch, auf Index 3 im String "hey" zuzugreifen
- Werte eines falschen Typs, z. B. Strings, wenn Ganzzahlen erwartet werden

---

*Input validation is the process of checking whether user input is valid before using it in a program. It helps prevent errors, unexpected behavior, and ensures the program receives data in the correct format or range.
Some examples include:*
- *missing or empty input values in mandatory fields, such as empty strings when the length of the string is critical*
- *negative values where only positive values are accepted, such as -15 as the amount of an ingredient in a recipe*
- *values that are too small or too large, for example when working with dates and times*
- *invalid indexes for sequences, such as trying to access index 3 in the string "hey"*
- *values of a wrong type, such as strings when integers are expected*

# Behandlung von Ausnahmen / *Exception Handling*

Fehler, die während der Ausführung eines Programms auftreten, werden Ausnahmen (Exceptions) genannt. Es ist möglich, sich auf Ausnahmen vorzubereiten und sie zu behandeln, sodass die Ausführung trotz ihres Auftretens fortgesetzt wird.

Die Ausnahmebehandlung in Python erfolgt mit den Anweisungen `try` und `except`. Die Idee ist, dass Python prüft, ob ein entsprechender except-Block vorhanden ist, wenn innerhalb eines try-Blocks eine Ausnahme ausgelöst wird. Existiert ein solcher Block, wird er ausgeführt und das Programm läuft dann weiter, als wäre nichts passiert.

| Block     | Zweck                                         |
| --------- | --------------------------------------------- |
| `try`     | Code, der einen Fehler verursachen könnte     |
| `except`  | Behandelt den Fehler                          |
| `else`    | Wird ausgeführt, wenn keine Ausnahme auftritt |
| `finally` | Wird unabhängig von Fehlern ausgeführt        |

---

*Errors that occur while the program is already running are called exceptions. It is possible to prepare for exceptions, and handle them so that the execution continues despite them occurring.*

*Exception handling in Python is accomplished with `try` and `except` statements. The idea is that if something within a try block causes an exception, Python checks if there is a corresponding except block. If such a block exists, it is executed and the program then continues as if nothing happened.*

| Block     | Purpose                        |
| --------- | ------------------------------ |
| `try`     | Code that might cause an error |
| `except`  | Handles the error              |
| `else`    | Runs if no exception occurs    |
| `finally` | Runs regardless of errors      |


````python
def divide(x, y):
    try:
        result = x / y
    except Exception as e:
        return f"Error: {e}"

    return result

print(divide(10, 2))
print(divide(10, 0))
````

````python
# rewrite divide with exception handling
def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        return "Cannot divide by zero"

    return result


print(divide(10, 2))
print(divide(10, 0))
````

````python
def add_numbers(a, b):
    try:
        return a + b
    except TypeError:
        return "Invalid types for addition"


print(add_numbers(3, 4))
print(add_numbers(3, "hello"))
````

````python
divide(3, "hello")
````

````python
def divide(a, b):
    result = None
    try:
        result = a / b
    except ZeroDivisionError:
        return "Division by zero error"
    except TypeError:
        return "Invalid types"
    except Exception as e:
        print(f"Something went wrong {e}")
    # excutes only when no error occurs
    else:
        print("Division operation is successfull.")
    # always executes
    finally:
        print("Division operation is concluded.")
    return result
````

````python
print(divide(10,2))
````

````python
print(divide(10,0))
````

````python
print(divide("10",2))
````

# Verschachteltes try-except / *Nested try-except*

Ermöglicht verschiedene Phasen eines Programms getrennt zu behandeln, insbesondere wenn:
- mehrere Operationen fehlschlagen könnten
- unterschiedliche Fehlertypen in verschiedenen Schritten auftreten
  
---

*It allows handling different stages of a program separately, especially when:*
- *multiple operations may fail*
- *different types of errors occur at different steps*

````python
def process_list(data):
    """
    Convert first 10 elements of the provided list to integers
    """
    result = []
    for index in range(10):
        try:
            value = data[index]
            try:
                number = int(value)
                print("Converted number:", number)
                result.append(number)
            except ValueError:
                print("Value cannot be converted to integer")

        except IndexError:
            print("Index is out of range")
    return result
````

````python
process_list(["10", "20", "abc"])
````

# Auslösen von Fehlern in Python / *Raising Errors in Python*

In Python können wir absichtlich einen Fehler auslösen, indem wir die raise-Anweisung verwenden, wenn das Programm einen ungültigen Zustand erkennt. Dies ist nützlich, wenn die Ausführung gestoppt werden soll und wirauf darauf hinweisen möchten, dass etwas schiefgelaufen ist, anstatt das Programm mit falschen Daten weiterlaufen zu lassen.

Syntax:
```
raise ExceptionType("Error message")
```

Das Auslösen von Fehlern hilft dabei:
- ungültige Eingaben frühzeitig zu erkennen
- falsches Programmverhalten zu verhindern
- klar zu kommunizieren, was schiefgelaufen ist
  
---

*In Python, you can raise an error intentionally using the `raise` statement when a program detects an invalid condition. This is useful when you want to stop execution and signal that something went wrong instead of allowing the program to continue with incorrect data.*

Syntax:
```
raise ExceptionType("Error message")
```

*Raising errors helps to:*
- *detect invalid input early*
- *prevent incorrect program behavior*
- *clearly communicate what went wrong*

````python
def withdraw(balance, amount):
    if amount > balance:
        raise ValueError("Insufficient balance")
    return balance - amount

print(withdraw(100, 50))
print(withdraw(100, 200))
````
